```python
import subprocess
import os


FILE_NAME = "./movie/movie.mp4"


def extract_audio_from_movie(movie_file_path):
    audio_file_path = movie_file_path[:-4] + ".wav"
    if os.path.exists(audio_file_path):
        os.remove(audio_file_path)
    result = subprocess.run(["ffmpeg", "-i", movie_file_path, audio_file_path], capture_output=True)
    if result.returncode == 0:
        print(f"success to extract audio data from {movie_file_path} to {audio_file_path}")
        return audio_file_path
    else: 
        print("fail")
        return False


audio_file_path = extract_audio_from_movie(FILE_NAME)
if audio_file_path is False:
    raise Exception
```

```python
import soundfile as sf
import os
import numpy as np
from matplotlib import pyplot as plt


def plot_audio_data(audio_file_path):
    data, samplerate = sf.read(audio_file_path, always_2d=True)
    os.remove(audio_file_path)
    t = np.arange(0, len(data))/samplerate # sec
    plt.figure(figsize=(18, 6))
    plt.plot(t, data)
    plt.show()
    return data, samplerate, t


data, samplerate, t = plot_audio_data(audio_file_path)
```

```python
# 音が一定量なっている期間だけTrue
def mark_over_threshold(data):
    thres = 0.02
    amp = np.abs(data)
    b = amp > thres
    return b

b = mark_over_threshold(data)
```

```python
# 2次元データで返ってくるので、片方でもなっているものはTrueにする
def reduce_audio_dimension(b):
    sounds_good = np.zeros(len(data))
    for i, v in enumerate(b):
        if v.all():
            sounds_good[i] = True
        else: 
            sounds_good[i] = False
    return sounds_good
b = reduce_audio_dimension(b)
```

```python
# 音が鳴っている部分だけplot
def plot_sounds_over_threshold(data, b, t):
    plt.figure(figsize=(18, 6))
    # plt.plot(t, b)
    plt.plot(t, b)
    plt.plot(t, abs(data))
    plt.show()
plot_sounds_over_threshold(data, b, t)
```

```python
min_silence_duration = 0.5


def mark_and_plot_silences(b, min_silence_duration):
    silences = []
    prev = 0
    entered = 0
    for i, v in enumerate(b):
        if prev == 1 and v == 0: # enter silence
            entered = i
        if prev == 0 and v == 1: # exit silence
            duration = (i - entered) / samplerate 
            if duration > min_silence_duration:
                silences.append({"from": entered, "to": i, "suffix": "cut"})
                entered = 0
        prev = v
    if entered > 0 and entered < len(b):
        silences.append({"from": entered, "to": len(b), "suffix": "cut"})

    plt.figure(figsize=(18, 6))
    cutpoint_from = np.zeros(len(data))
    cutpoint_to = np.zeros(len(data))
    for silence in silences:
        cutpoint_from[silence["from"]] = 1
        cutpoint_to[silence["to"]-1] = 1
    plt.plot(t, cutpoint_from)
    plt.plot(t, cutpoint_to)
    plt.plot(t, abs(data))
    plt.show()
    return silences
    
    
silences = mark_and_plot_silences(b, min_silence_duration)
```

```python
# 瞬間的に音が鳴っている部分をカット

min_keep_duration = 0.3
def denoise(min_keep_duration, silences):
    cut_blocks = []
    blocks = silences.copy()
    while 1:
        if len(blocks) == 0:
            break
        if len(blocks) == 1:
            cut_blocks.append(blocks[0])
            break
        block = blocks[0]
        next_blocks = [block]
        for i, b in enumerate(blocks):
            if i == 0:
                continue
            interval = (b["from"] - block["to"]) / samplerate 
            if interval < min_keep_duration:
                block["to"] = b["to"]
                next_blocks.append(b)
        cut_blocks.append(block)
        blocks = list(filter(lambda b: b not in next_blocks, blocks))
    plt.figure(figsize=(18, 6))
    cutpoint_from = np.zeros(len(data))
    cutpoint_to = np.zeros(len(data))
    for silence in cut_blocks:
        cutpoint_from[silence["from"]] = 1
        cutpoint_to[silence["to"]-1] = 1
    plt.plot(t, cutpoint_from)
    plt.plot(t, cutpoint_to)
    plt.plot(t, abs(data))
    return cut_blocks
        
cut_blocks = denoise(min_keep_duration, silences)
```

```python
# カット部分の反転
def make_keep_blocks(cut_blocks):
    keep_blocks = []
    for i, block in enumerate(cut_blocks):
        if i == 0 and block["from"] > 0:
            keep_blocks.append({"from": 0, "to": block["from"], "suffix": "keep"})
        if i > 0:
            prev = cut_blocks[i - 1]
            keep_blocks.append({"from": prev["to"], "to": block["from"], "suffix": "keep"})
        if i == len(cut_blocks) - 1 and block["to"] < len(data):
            keep_blocks.append({"from": block["to"], "to": len(data), "suffix": "keep"})
    return keep_blocks

keep_blocks = make_keep_blocks(cut_blocks)
```

```python
import time
import subprocess

padding_time = 0.1
def jump_cut(FILE_NAME, keep_blocks):
    file_list = []
    file_list_path = "file_list.txt"
    for i, block in enumerate(keep_blocks):
        fr = max(block["from"] / samplerate - padding_time, 0)
        to = min(block["to"] / samplerate + padding_time, len(data) / samplerate)
        duration = to - fr
        out_path = os.path.join("{}_{}.mov".format(i, block["suffix"]))
        file_list.append("file " + out_path + "\n")
        result = subprocess.run(["ffmpeg", "-ss", str(fr), "-i", str(FILE_NAME), "-t", str(duration), str(out_path)], capture_output=True)
    f = open(file_list_path, "a")
    f.writelines(file_list)
    f.close()
    # ffmpeg -f concat -i mylist.txt -c copy output.MOV
    subprocess.run(["ffmpeg", "-f", "concat", "-i", file_list_path, "-c", "copy", "movie\output.MOV"])

    for file in file_list:
        os.remove(file.replace("file ", "").replace("\n", ""))
    os.remove(file_list_path)
    
jump_cut(FILE_NAME, keep_blocks)
```
